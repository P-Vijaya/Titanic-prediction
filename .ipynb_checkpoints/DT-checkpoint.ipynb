```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
from pandas import Series,DataFrame
#from Pylab import rcParams
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import StandardScaler
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pickle
```

```python
titanic = pd.read_csv("https://raw.githubusercontent.com/BigDataGal/Python-for-Data-Science/master/titanic-train.csv")
titanic
```

```python
data = titanic[['Pclass','Sex','Age','SibSp','Parch','Fare','Survived']]
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Age'].hist()
```

```python
data['Age_pfill'] = data['Age'].fillna(method='pad')
```

```python
data['Age_pfill'].hist()
```

```python
## dropping Age column
data.drop('Age',axis=1,inplace=True)
```

```python
data.head()
```

```python
## Now there is no missing values
data.isnull().sum()
```

```python
## encoding categorical columns

data['Sex']=np.where(data['Sex']=='male',1,0)
data.head()
```

```python
X = data.drop(columns='Survived')
y = data['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
y_pred = dt.predict(X_test)
```

```python
## for train data
dt.score(X_train,y_train)
```

```python
## for test data
dt.score(X_test,y_test)
```

```python
## accuracy score
metrics.accuracy_score(y_test,y_pred)
```

```python
plt.figure(figsize=(25,30))
tree.plot_tree(dt,filled=True)
```

```python
std_scalar = StandardScaler()
X_scaled = std_scalar.fit_transform(X)
```

```python
### Multicollinearity
variables = X_scaled
vif = DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(variables,i) for i in range(variables.shape[1])]
vif
```

All the values are below 5 hence there is no multicollinearity between features.

```python
from sklearn.decomposition import PCA
import numpy as np
pca = PCA()
principalComponents = pca.fit_transform(X_scaled)
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Explained Variance')
plt.show()
```

We can see that around 95% of the variance is being explained by 4 components.So instead of giving all 6 columns as input in our algorithm let's use these 4 principal components instead.

```python
pca = PCA(n_components=4)
new_data = pca.fit_transform(X_scaled)
principal_x = pd.DataFrame(new_data,columns=['PC-1','PC-2','PC-3','PC-4'])
```

```python
principal_x
```

```python
X_train,X_test,y_train,y_test = train_test_split(principal_x,y,test_size=0.25,random_state=1)
```

```python
dt1 = DecisionTreeClassifier()
dt1.fit(X_train,y_train)
y_pred1 = dt1.predict(X_test)
metrics.accuracy_score(y_test,y_pred1)
```

```python
print("Pricncipal train:",dt1.score(X_train,y_train))
print("Principal test",dt1.score(X_test,y_test))
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.25,random_state=1)
```

```python
dt2 = DecisionTreeClassifier()
dt2.fit(X_train,y_train)
y_pred2 = dt2.predict(X_test)
metrics.accuracy_score(y_test,y_pred2)
```

```python
print("Scaled train:",dt2.score(X_train,y_train))
print("Scaled test",dt2.score(X_test,y_test))
```

Analysis:

We can see that,

- The accuracy score of imputing the missing and encoding catg features is 77%.
- The accuracy score of scaling using standard scaling is 77%.
- Whereas, the accuracy score using principal components is only 73%.

Therefore choosing the data with the imputing,encoding and scaled data ie)dt2.

### Hyperparameter tuning

```python
grid_param = {
    'criterion': ['gini', 'entropy'],
    'splitter' : ['best', 'random'],
    'max_depth' : range(2,32,1),
    'min_samples_leaf' : range(1,10,1),
    'min_samples_split': range(2,10,1),
}
```

```python
grid_search = GridSearchCV(estimator=dt2,param_grid=grid_param,cv=5,n_jobs=-1,verbose=1)
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
dt3 = DecisionTreeClassifier(criterion='entropy',max_depth= 5,min_samples_leaf= 1,min_samples_split= 3,splitter='best')
dt3.fit(X_train,y_train)
y_pred3 = dt3.predict(X_test)
metrics.accuracy_score(y_test,y_pred3)
```

```python
print("GS train:",dt3.score(X_train,y_train))
print("GS test",dt3.score(X_test,y_test))
```

```python
random_search = RandomizedSearchCV(estimator=dt2,param_distributions=grid_param,cv=10,n_jobs=-1,verbose=1)
random_search.fit(X_train,y_train)
```

```python
random_search.best_params_
```

```python
dt4 = DecisionTreeClassifier(criterion='entropy',max_depth= 5,min_samples_leaf= 4,min_samples_split= 2,splitter='best')
dt4.fit(X_train,y_train)
y_pred4 = dt4.predict(X_test)
metrics.accuracy_score(y_test,y_pred4)
```

```python
print("RS train:",dt4.score(X_train,y_train))
print("RS test",dt4.score(X_test,y_test))
```

```python
dt5 = DecisionTreeClassifier(criterion='entropy',max_depth=6,min_samples_leaf=7,min_samples_split=25,splitter='best')
dt5.fit(X_train,y_train)
y_pred5 = dt5.predict(X_test)
metrics.accuracy_score(y_test,y_pred5)
```

```python
print("RS_new train:",dt5.score(X_train,y_train))
print("RS_new test",dt5.score(X_test,y_test))
```

dt5 model is considered as final model, as we achieved an accuracy score of 84% from 77%.

```python
## saving the model in local file system
filename = 'dt.pickle'
pickle.dump(dt5,open(filename,'wb'))
```
